```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
train = pd.read_csv('./kaggle/input/train.csv', index_col='PassengerId') #read data from train.csv
test = pd.read_csv('./kaggle/input/test.csv', index_col='PassengerId') #read data from test.csv
```

```python
head = train[:100]
head #the first 100 rows in train data
```

```python
middle_first_class_age,middle_first_class_fare = train.loc[train['Pclass']==1].Age.mean(), train.loc[train['Pclass']==1].Fare.mean()
middle_second_class_age,middle_second_class_fare = train.loc[train['Pclass']==2].Age.mean(), train.loc[train['Pclass']==2].Fare.mean()
middle_third_class_age,middle_third_class_fare = train.loc[train['Pclass']==3].Age.mean(), train.loc[train['Pclass']==3].Fare.mean()
```

```python
train['Age'] = np.where(((train.Pclass == 1) & (pd.isna(train['Age']))), middle_first_class_age, train.Age)
train['Age'] = np.where(((train.Pclass == 2) & (pd.isna(train['Age']))), middle_second_class_age, train.Age)
train['Age'] = np.where(((train.Pclass == 3) &( pd.isna(train['Age']))), middle_third_class_age, train.Age)

test['Age'] = np.where(((test.Pclass == 1) & (pd.isna(test['Age']))), middle_first_class_age, test.Age)
test['Age'] = np.where(((test.Pclass == 2) & (pd.isna(test['Age']))), middle_second_class_age, test.Age)
test['Age'] = np.where(((test.Pclass == 3) &( pd.isna(test['Age']))), middle_third_class_age, test.Age)
test['Fare'] = test['Fare'].fillna(value=test['Fare'].mean())
```

```python
from sklearn.neighbors import KNeighborsClassifier

features = train[['Age', 'Fare']].copy()
features_test = test[['Age', 'Fare']].copy()
answers = train["Survived"]

model = KNeighborsClassifier(n_neighbors=9)
model.fit(features, answers)

test_predictions = model.predict(features_test)

output = pd.DataFrame({'PassengerId': test.index,
                      'Survived': test_predictions})
output.to_csv('./kaggle/output/submission.csv', index=False)
```
